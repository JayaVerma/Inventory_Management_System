```python

import json

fd = open("records.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

```python
def thankyouNote():
    print()
    print('*********************************')
    print("Thank you for visiting the inventory")
```

```python
def addItemsToInventary():
    print("WELCOME TO ADD ITEMS TO INVENTORY")
    print("---------------------------------")
    print()
    print("Available Options to Add:")
    print("1 - Add Item to existing Product Id")
    print("2 - Add new product id with details to the inventory ")
    print()
    while True:
        userChoice = input("Choose An Option: ")
        if userChoice == '1':
            productid = input("Enter the product id: ")
            quantity = int(input("Enter the new adding quantity: "))
            records[productid]["Quantity"] = records[productid]["Quantity"] + quantity
            print(records[productid])
            print()
            print("The items to is updated to the Inventory")
            break
        elif userChoice == '2':
            newProductid = input("Enter the product id: ")
            newProductname = input("Enter the product name: ")
            newProductPrice = int(input("Enter the product Price: "))
            newProductCategory = input("Enter the product Category: ")
            newProductQuantity = int(input("Enter the product Quantity: "))
            newProductDurability = input("Enter the product Durability: ")
            records[newProductid] = {"ProductName":newProductname, "ProductPrice":newProductPrice, 
                                     "ProductCategory":newProductCategory, "ProductQuantity":newProductQuantity,
                                     "ProductDurability":newProductDurability
                                    }
            print()
            print("The New product is added successfully")
            print()
            
            break
        else:
            print('Invalid Input')
    js = json.dumps(records)
    fd = open("records.json",'w')
    fd.write(js)
    fd.close()
    print("The New product is updated to The Inventory successfully")
    thankyouNote()
```

```python
def deleteInventoryItems():
    print("WELCOME TO DELETING ITEM FROM INVENTORY")
    print("----------------------------------------")
    print()
    productid = input("Enter the product Id you want to delete: ")
    if productid in records:
        opt3 = input("Are you sure You want to delete the complete product - ")
        if opt3 == 'Yes':
            removedProduct = records.pop(productid)
            print()
            print("The removed Product is: " + str(removedProduct))
            print()
            js = json.dumps(records)
            fd = open("records.json",'w')
            fd.write(js)
            fd.close()
            print("The Inventory is Updated")
            
    else:
        print("Invalid product id")
    thankyouNote()    
```

```python
def purchaseItemFromInventory():
    
    print("WELCOME TO PURCHASE ITEM FROM INVENTORY")
    print("----------------------------------------")
    print()
    num = int(input('Enter number of different products you want to buy - '))
    billAmount = []
    
    for i in range(num):
        sales = {}
        productid = input("Enter the product id: ")
        quantity = int(input("Enter quantity you want to buy: "))
        if productid in records:
            print(records[productid])
            records[productid]["Quantity"] = records[productid]["Quantity"] - quantity
            Amount = records[productid]["ProductPrice"]*quantity
            billAmount.append(Amount)
            saledItem = {"ProductName":records[productid]["ProductName"], 
                         "ProductPrice":records[productid]["ProductPrice"], 
                         "ProductCategory":records[productid]["Category"], 
                         "ProductQuantity":quantity,
                         "ProductDurability":records[productid]["Durability"] }  
            sales.update(saledItem)
            mysale = json.dumps(sales)
            saleRecord = open("SalesRecord.json",'a')
            saleRecord.write(mysale)
            saleRecord.close()
            
    print("-------------------------------------")   
    print("total Billing Amount: " + str(sum(billAmount)))
    print()
    print("Do you want to buy Something else ?")
    opt2 = input("Yes or No - ")
    if opt2=='Yes':
        purchaseItemFromInventory()
    else:
        thankyouNote()
        
    js = json.dumps(records)
    fd = open("records.json",'w')
    fd.write(js)
    fd.close()
    
    
```

```python
def UpdateItemsPrice():
    print("WELCOME TO UPDATING PRICE OF THE ITEM IN INVENTORY")
    print("---------------------------------------------------")
    print()
    productid = input("Enter the product id: ")
    newPrice = int(input("Enter new price of the product: "))
    records[productid]["ProductPrice"] = newPrice
    
    js = json.dumps(records)
    fd = open("records.json",'w')
    fd.write(js)
    fd.close()
```

```python
def veiwItemsInInventory():
    print("WELCOME TO VIEW INVENTORY")
    print("-------------------------")
    print()
    print("Name of all the products available in the inventory: ")
    for productid in records:
        productName = records[productid]["ProductName"]
        print(str(productid) + ":" + str(productName))
        
    print()   
    print("Do you want to buy Something?")
    opt2 = input("Yes or No - ")
    if opt2=='Yes':
        purchaseItemFromInventory()
    else:
        thankyouNote()       
```
